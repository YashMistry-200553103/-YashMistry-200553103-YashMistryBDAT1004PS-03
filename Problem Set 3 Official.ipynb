# Question 1

### Step 1. Import the necessary libraries 
### Step 2. Import the dataset from this address. 
### Step 3. Assign it to a variable called users 

```python
import pandas as pd

URLlink = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
Users = pd.read_csv(URLlink, sep='|')

print("Users_Data:")
print(Users)

tabulated_output = Users.to_markdown(index=False)
print("\nUsers_Data:")
print(tabulated_output)
```

### Step 4. Discover what is the mean age per occupation 

```python
mean_age_per_occupation = Users.groupby('occupation')['age'].mean().reset_index()



tabulated_output = mean_age_per_occupation.to_markdown(index=False)
print("\nmean_age_per_occupation:")
print(tabulated_output)
```

### Step 5: Discover the Male ratio per occupation and sort it from the most to the least

```python
Users['Male_Ratio'] = (Users['gender'] == 'M').astype(int)

Male_Ratio_by_Occupation = Users.groupby('occupation')['Male_Ratio'].mean().sort_values(ascending=False)

tabulated_output = Male_Ratio_by_Occupation.reset_index().to_markdown(index=False)
print("\nMale_Ratio_by_Occupation:")
print(tabulated_output)
```

### Step 6:. For each occupation, calculate the minimum and maximum ages

```python
Age_Ranges_by_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

tabulated_output = Age_Ranges_by_occupation.reset_index().to_markdown(index=False)
print("\nAge_Ranges_by_occupation:")
print(tabulated_output)
```

### Step 7: For each combination of occupation and sex, calculate the mean age

```python
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()

tabulated_output = mean_age_by_occupation_sex.reset_index().to_markdown(index=False)
print("\nMean age for each combination of occupation and sex:")
print(tabulated_output)
```

### Step 8: For each occupation present the percentage of women and men

```python
Gender_count_by_occupation = users.groupby('occupation')['gender'].value_counts()

Gender_total_by_occupation = users.groupby('occupation')['gender'].count()

Gender_percentage_by_occupation = Gender_count_by_occupation.div(Gender_total_by_occupation, level='occupation') * 100

Gender_percentage_by_occupation = Gender_percentage_by_occupation.unstack().reset_index()

Gender_percentage_by_occupation.columns = ['occupation', 'Female', 'Male']

tabulated_output = Gender_percentage_by_occupation.to_markdown(index=False)
print("\nPercentage of women and men for each occupation:")
print(tabulated_output)
```

```python
import pandas as pd

# Step 2: Import the dataset
file_location = r"C:\Users\Admin\Desktop\Sem 1\BDAT 1004 D.P Roshan sahu\Assignment\Problem Set 3\wind.txt"
data = pd.read_csv(file_location, delim_whitespace=True, parse_dates=[[0, 1, 2]])

def fix_year(year):
    if year.year > 1989:
        return year - pd.DateOffset(years=100)
    return year

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)

# Step 5: Set the right dates as the index
data.set_index('Yr_Mo_Dy', inplace=True)

# Step 6: Count missing values for each location
missing_values_per_location = data.isnull().sum()

# Step 7: Count non-missing values in total
total_non_missing_values = data.notnull().sum().sum()

# Step 8: Calculate the mean windspeeds over all the locations and times
mean_windspeed = data.mean().mean()

# Step 9: Calculate statistics for each location
loc_stats = data.groupby(data.index.to_period('A')).describe()

# Step 10: Calculate statistics for each day
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])

# Step 11: Average windspeed in January for each location
january_avg_per_location = data.resample('M').mean()
january_avg_per_location = january_avg_per_location[january_avg_per_location.index.month == 1]

# Step 12: Downsample to yearly frequency
yearly_data = data.resample('A').mean()

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W-MON').mean()  # Assume the first week starts on Monday, January 2, 1961

# Step 15: Calculate statistics for each week (first 52 weeks)
weekly_stats = weekly_data.head(52).agg(['min', 'max', 'mean', 'std'])

print("Missing Values:")
print(missing_values_per_location)
print("\nNon-Missing Count:")
print(total_non_missing_values)
print("\nMean Windspeed over all locations and times:")
print(mean_windspeed)
print("\nLocation Statistics:")
print(loc_stats)
print("\nDay Statistics:")
print(day_stats)
print("\nAverage Windspeed in January:")
print(january_avg_per_location)
print("\nYearly Frequency:")
print(yearly_data)
print("\nMonthly Frequency:")
print(monthly_data)
print("\nWeekly Frequency for the first 52 weeks:")
print(weekly_stats)
```

```python
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print("First 10 Entries")
print(chipo.head(10))

# Step 5: Number of observations
num_observations = chipo.shape[0]
print("Number of Observations:", num_observations)

# Step 6: Number of columns
num_columns = chipo.shape[1]
print("Total Number of Columns in Chipo dataset:", num_columns)

# Step 7: Print column names
print("Chipo Dataset Column Names:", chipo.columns.tolist())

# Step 8: Indexing
print("Dataset Indexing:", chipo.index)

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("Most-Ordered Item of Chipo:", most_ordered_item)

# Step 10: Quantity of most-ordered item
quantity_most_ordered_item = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("Most-Ordered Item Quantity:", quantity_most_ordered_item)

# Step 11: Most-ordered item in choice_description column
most_ordered_choice = chipo['choice_description'].value_counts().idxmax()
print("Most-Ordered Item from the Choice Description:", most_ordered_choice)

# Step 12: Total items ordered
total_items_ordered = chipo['quantity'].sum()
print("Total number of Items Ordered:", total_items_ordered)

# Step 13: Convert item price to float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Step 14: Revenue
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Total Revenue from the dataset:", revenue)

# Step 15: Number of orders
num_orders = chipo['order_id'].nunique()
print("Number of Orders made in the period of dataset:", num_orders)

# Step 16: Average revenue per order
avg_revenue_per_order = revenue / num_orders
print("Average Revenue amount per Order:", avg_revenue_per_order)

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()
print("Number of Different Items Sold:", num_different_items)
```
